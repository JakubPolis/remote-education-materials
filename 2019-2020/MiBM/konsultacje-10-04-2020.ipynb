Znaleźć macierz odwrotną do macierzy $\begin{bmatrix}
    1 & 2 & 3\\
    2 & 3 & 7\\
    0 & 0 & 1
\end{bmatrix}$.

$
\begin{bmatrix}
    1 & 2 & 3 & | & 1 & 0 & 0\\
    2 & 3 & 7 & | & 0 & 1 & 0\\
    0 & 0 & 1 & | & 0 & 0 & 1
\end{bmatrix}
\overset{w_2 - 2 w_1}{\to}
\begin{bmatrix}
    1 & 2 & 3 & | & 1 & 0 & 0\\
    0 & -1 & 1 & | & -2 & 1 & 0\\
    0 & 0 & 1 & | & 0 & 0 & 1
\end{bmatrix}
\overset{w_2 * (-1)}{\to}
$

$
\begin{bmatrix}
    1 & 2 & 3 & | & 1 & 0 & 0\\
    0 & 1 & -1 & | & 2 & -1 & 0\\
    0 & 0 & 1 & | & 0 & 0 & 1
\end{bmatrix}
\overset{w_1 - 3 * w_3 \\ w_2 + w_3}{\to}
\begin{bmatrix}
    1 & 2 & 0 & | & 1 & 0 & -3\\
    0 & 1 & 0 & | & 2 & -1 & 1\\
    0 & 0 & 1 & | & 0 & 0 & 1
\end{bmatrix}
\overset{w_1 - 2 * w_2}{\to}
$

$
\begin{bmatrix}
    1 & 0 & 0 & | & -3 & 2 & -5\\
    0 & 1 & 0 & | & 2 & -1 & 1\\
    0 & 0 & 1 & | & 0 & 0 & 1
\end{bmatrix}
$

**Odp.:** Macierz odwrotna do danej macierzy to $\begin{bmatrix}
    -3 & 2 & -5\\
    2 & -1 & 1\\
    0 & 0 & 1
\end{bmatrix}$.

```python
import numpy

A = numpy.array([[1,2,3],[2,3,7],[0,0,1]])
print(A)
print()
B = numpy.array([[-3,2,-5],[2,-1,1],[0,0,1]])
print(B)
print()

print(numpy.dot(A,B))
print()
print(numpy.dot(B,A))
print()
```

Znaleźć macierz odwrotną do macierzy $\begin{bmatrix}
    0 & 1 & 2 & 4\\
    2 & 3 & 5 & 9\\
    3 & 5 & 8 & 13\\
    3 & 5 & 7 & 11
\end{bmatrix}$.

$
\begin{bmatrix}
    0 & 1 & 2 & 4 & | & 1 & 0 & 0 & 0\\

    3 & 5 & 8 & 13 & | & 0 & 0 & 1 & 0\\
    3 & 5 & 7 & 11 & | & 0 & 0 & 0 & 1
\end{bmatrix}
\overset{w_3 - w_2\\ w_4 - w_2}{\to}
\begin{bmatrix}
    0 & 1 & 2 & 4 & | & 1 & 0 & 0 & 0\\
    2 & 3 & 5 & 9 & | & 0 & 1 & 0 & 0\\
    1 & 2 & 3 & 4 & | & 0 & -1 & 1 & 0\\
    1 & 2 & 2 & 2 & | & 0 & -1 & 0 & 1
\end{bmatrix}
\overset{w_1 \leftrightarrow w_4}{\to}
$

$
\begin{bmatrix}
    1 & 2 & 2 & 2 & | & 0 & -1 & 0 & 1\\

    1 & 2 & 3 & 4 & | & 0 & -1 & 1 & 0\\
    0 & 1 & 2 & 4 & | & 1 & 0 & 0 & 0
\end{bmatrix}
\overset{w_2 - 2 * w_1\\ w_3 - w_1}{\to}
\begin{bmatrix}
    1 & 2 & 2 & 2 & | & 0 & -1 & 0 & 1\\
    0 & -1 & 1 & 5 & | & 0 & 3 & 0 & -2\\
    0 & 0 & 1 & 2 & | & 0 & 0 & 1 & -1\\
    0 & 1 & 2 & 4 & | & 1 & 0 & 0 & 0
\end{bmatrix}
\overset{w_4 + w_2}{\to}
$

$
\begin{bmatrix}

    0 & -1 & 1 & 5 & | & 0 & 3 & 0 & -2\\
    0 & 0 & 1 & 2 & | & 0 & 0 & 1 & -1\\

\overset{w_4 - 3 * w_3}{\to}
\begin{bmatrix}

    0 & -1 & 1 & 5 & | & 0 & 3 & 0 & -2\\
    0 & 0 & 1 & 2 & | & 0 & 0 & 1 & -1\\
    0 & 0 & 0 & 3 & | & 1 & 3 & -3 & 1
\end{bmatrix}
\overset{w2 * (-1)\\ w_4 * \frac{1}{3}}{\to}
$

$
\begin{bmatrix}
    1 & 2 & 2 & 2 & | & 0 & -1 & 0 & 1\\
    0 & 1 & -1 & -5 & | & 0 & -3 & 0 & 2\\
    0 & 0 & 1 & 2 & | & 0 & 0 & 1 & -1\\
    0 & 0 & 0 & 1 & | & \frac{1}{3} & 1 & -1 & \frac{1}{3}
\end{bmatrix}
\overset{w_2 + w_3}{\to}
\begin{bmatrix}
    1 & 2 & 2 & 2 & | & 0 & -1 & 0 & 1\\
    0 & 1 & 0 & -3 & | & 0 & -3 & 1 & 1\\
    0 & 0 & 1 & 2 & | & 0 & 0 & 1 & -1\\
    0 & 0 & 0 & 1 & | & \frac{1}{3} & 1 & -1 & \frac{1}{3}
\end{bmatrix}
\overset{w_1 - 2 * w_4\\ w_2 + 3*w_4\\ w_3 - 2 * w_4}{\to}
$

$
\begin{bmatrix}
    1 & 2 & 2 & 0 & | & \frac{-2}{3} & -3 & 2 & \frac{1}{3}\\
    0 & 1 & 0 & 0 & | & 1 & 0 & -2 & 2\\

\overset{w_1 - 2* w_2\\ w_1 - 2*w_3}{\to}
\begin{bmatrix}

    0 & 1 & 0 & 0 & | & 1 & 0 & -2 & 2\\

**Odp.:** Macierz odwrotna do danej macierzy to $\begin{bmatrix}
